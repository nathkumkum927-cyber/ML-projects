```python
filtered_df = df[(df['petal_length'] > 4.5) & (df['species'] == 'Iris-virginica')]
print(filtered_df)
```

```python
result = df.groupby('species').agg(
    avg_sepal_length=('sepal_length', 'mean'),
    max_petal_width=('petal_width', 'max'),
    std_sepal_width=('sepal_width', 'std')
)

print(result)
```

```python
df['petal_ratio'] = df['petal_length'] / df['petal_width']
```

```python
avg_petal_ratio = df.groupby('species')['petal_ratio'].mean()
print(avg_petal_ratio)
```

```python
df.groupby('species')['petal_length'].mean() / df.groupby('species')['petal_width'].mean()
```
